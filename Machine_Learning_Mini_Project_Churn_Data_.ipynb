```python
data.drop(['customerID','gender','PaymentMethod','Contract'],axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lblEnc=LabelEncoder()
data_copy = data.copy()

for i in data.columns:
  data_copy[i]=lblEnc.fit_transform(data[i])
```

```python
data_copy
```

```python
X=data_copy.drop("Churn",axis = 1)
Y=data_copy["Churn"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,"Traning Feature Shape")
print(X_test.shape,"Testing Feature Shape")
print(y_train.shape,"Traning Labels Shape")
print(y_test.shape,"Testing Labels Shape")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data_log_model=LogisticRegression()
```

```python
data_log_model.fit(X_train,y_train)
```

```python
data_predictions = data_log_model.predict(X_test)
```

```python
data_predictions 
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,data_predictions))
```

```python
import sklearn.metrics as sm 
```

```python
accuracy_data_model =sm.accuracy_score(y_test,data_predictions)
print("Accuracy Score ===",str(accuracy_data_model))
```
